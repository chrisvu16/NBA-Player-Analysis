```r
# Load necessary libraries
library(glmnet)
pointGuard$sqrtreb = sqrt(pointGuard$rebounds)
pointGuard$sqrtass=sqrt(pointGuard$assists)
# Define the formula
#formula <- points_scored ~ pos + hdiff + wdiff + pheight + pwingspan
#formula <- sqrtreb ~  pos + wdiff + pheight + pwingspan + pweight
formula=sqrtass~pos+wdiff
# Prepare training data for regularization
x <- model.matrix(formula, pointGuard)[, -1]  # Remove intercept column
#y <- pointGuard$points_scored
#y=pointGuard$sqrtreb
y=pointGuard$sqrtass
ridge_cv <- glmnet(x, y, alpha = 0)
lasso_cv <- glmnet(x, y, alpha = 1)

ridge_lambda <- ridge_cv$lambda.min
lasso_lambda <- lasso_cv$lambda.min

ridge_pred <- predict(ridge_cv, s = ridge_lambda, newx = x)
lasso_pred <- predict(lasso_cv, s = lasso_lambda, newx = x)
mse_ridge <- mean((y - ridge_pred)^2)
mse_lasso <- mean((y - lasso_pred)^2)

residuals_ridge <- y - ridge_pred
plot(ridge_pred, residuals_ridge)
abline(h = 0, col = "red")

qqnorm(residuals_ridge, main = "Q-Q Plot for Ridge Regression Residuals")
qqline(residuals_ridge, col = "red")

residuals_lasso <- y - lasso_pred
plot(lasso_pred, residuals_lasso)
abline(h = 0, col = "red")

qqnorm(residuals_lasso, main = "Q-Q Plot for Lasso Regression Residuals")
qqline(residuals_lasso, col = "red")

# Plot coefficient paths
plot(ridge_cv, xvar = "lambda", label = TRUE,
     main = "Ridge Regression Coefficient Paths")

```

```r

# Fit Ridge Regression
ridge_model <- glmnet(x, y, alpha = 0)

# Extract residual deviance
ridge_deviance <- ridge_model$dev.ratio * ridge_model$nulldev

# Number of parameters (including intercept)
num_params <- ridge_model$df + 1

# Calculate AIC for each lambda
aic <- 2 * num_params - 2 * log(ridge_deviance)
print(min(aic))
# Find the lambda with minimum AIC
best_lambda <- ridge_model$lambda[which.min(aic)]
print(best_lambda)
```
